```python
from ultralytics import YOLO

model = YOLO('/kaggle/working/inputdata/weights/yolov8n.pt')

full_model = model.model.model
for i, layer in enumerate(full_model):
    print(f"Layer {i}: {type(layer).__name__}")
```

```python
import math
#Recommended freeze ratio for a medium sized dataset like ours with 5385 examples in the training dataset is 0.2-0.4
min_freeze = math.floor(len(full_model) * 0.2)
max_freeze = math.ceil(len(full_model) * 0.4)

for num_layers_to_freeze in range(min_freeze, max_freeze):
    for i, layer in enumerate(full_model):
        if i < num_layers_to_freeze:
            for param in layer.parameters():
                param.requires_grad = False
        else:
            for param in layer.parameters():
                param.requires_grad = True

    model.train(
        data='/kaggle/working/inputdata/yolov8.yaml',
        epochs=100,
        imgsz=416,
        batch=32,
        name='train_with_freeze_' + str(num_layers_to_freeze) + "_layers",
        patience=10,
        verbose=True
    )
```

```python
#best performing model from training was the yolov8 model with the first 4 layers frozen
model = YOLO("/kaggle/working/inputdata/runs/detect/train_with_freeze_4_layers/weights/best.pt")

results = model.predict(
    source="/kaggle/working/inputdata/dataset/images/test",
    conf=0.225,                     # confidence threshold based on peak of f1_curve
    save=True,                     
    save_txt=True,                 
    project="runs/detect/test",        
    name="test_results",           
    imgsz=416                      
)

model.val(
    data="/kaggle/working/inputdata/yolov8.yaml",   
    split="test",         
    imgsz=416             
)
```
