```python
from ultralytics import YOLO

model = YOLO('/kaggle/working/inputdata/weights/yolo11n.pt')

model.train(
    data='/kaggle/working/inputdata/yolov11.yaml',
    epochs=100,
    imgsz=416,
    batch=32,
    name='train',
    patience=10,
    verbose=True
)
```

```python
model = YOLO("/kaggle/working/inputdata/runs/detect/train/weights/best.pt")

results = model.predict(
    source="/kaggle/working/inputdata/dataset/images/test",
    conf=0.208,                     # confidence threshold based on peak of f1_curve
    save=True,                     
    save_txt=True,                 
    project="runs/detect/test",        
    name="test_results",           
    imgsz=416                      
)

model.val(
    data="/kaggle/working/inputdata/yolov11.yaml",   
    split="test",         
    imgsz=416             
)
```
